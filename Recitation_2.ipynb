# Apply Function

```python
def compute(x):
      if(x > 60):
        return "Old"
      else:
        return "Young"

df['Old_Young'] = df.age.apply(compute)
```

```python
df['Old_Young']
```

# Group By Function

```python
df[df['sex'] == 1]['age']
```

```python
df[df['sex'] == 1]['age'].mean()
```

```python
df[df['sex'] == 2]['age']
```

```python
df[df['sex'] == 2]['age'].mean()
```

```python
df.groupby('sex')[['age']].mean()
```

```python
df.groupby('sex')[['age']].median()
```

```python
df.groupby(['sex', 'big5a1'])[['age']].mean()
```

```python
df.groupby(['sex', 'big5a1']).agg({'age': ['mean', 'min', 'max'], 'hrs1': ['min', 'max', 'median']})
```

# CrossTab Function

```python
pd.crosstab(index=df["sex"], columns=df["big5a1"])
```

```python
pd.crosstab(index=df["sex"], columns=df["big5a1"], values=df["age"], aggfunc = ["mean", "max", "min"])
```

```python
pd.crosstab(columns=df["sex"], index=df["big5a1"], values=df["age"], aggfunc = ["mean", "max", "min"])
```
